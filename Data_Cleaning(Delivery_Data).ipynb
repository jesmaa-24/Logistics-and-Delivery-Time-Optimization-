```python
import pandas as pd

df = pd.read_csv("Delivery_Logistics.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
df.describe()
```

```python
df['delivery_time_hours'] = pd.to_datetime(df['delivery_time_hours']).dt.nanosecond
df['expected_time_hours'] = pd.to_datetime(df['expected_time_hours']).dt.nanosecond
```

```python
df[['delivery_time_hours','expected_time_hours']].head()
```

```python
df['delivery_difference'] = df['delivery_time_hours'] - df['expected_time_hours']
```

```python
df['Calculated_delay_flag'] = df['delivery_difference'].apply(lambda x: 'Late' if x>0 else 'On-Time')
```

```python
df = df.dropna()
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.head()
df.info()
```

```python
df.to_csv("Clean_Delivery_Data.csv", index=False)
```
